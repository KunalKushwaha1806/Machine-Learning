# Program 8
### **Linear Regression (from Scratch)**

---

### (1) : Model (Hypothesis)

$$
\hat{y}_i = m x_i + b
$$


### (2) : Cost Function (Mean Squared Error)

$$
J(m,b) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

---

### (3) : Partial Derivative with respect to $m$ and $b$)

$$
\frac{\partial J}{\partial m}
=
-\frac{2}{n}
\sum_{i=1}^{n}
x_i (y_i - \hat{y}_i)
$$

$$
\frac{\partial J}{\partial b}
=
-\frac{2}{n}
\sum_{i=1}^{n}
(y_i - \hat{y}_i)
$$

---

### (4): Gradient Descent Update Rule

$$
m := m + \alpha \frac{2}{n}
\sum_{i=1}^{n}
x_i (y_i - \hat{y}_i)
$$

$$
b := b + \alpha \frac{2}{n}
\sum_{i=1}^{n}
(y_i - \hat{y}_i)
$$

---

```python
#Step 1: Model Hypothesis (Prediction)

def predict(x, m, b):
    return m * x + b
```

```python
#Step 2: Cost function (MSE)

def compute_cost(y, y_hat):
    n = len(y)
    return (1/n) * ((y - y_hat) ** 2).sum()
```

```python
#Step 3: Partial derivative with respect to m and b

def compute_gradients(x, y, y_hat):
    n = len(x)
    error = y - y_hat

    dm = (-2/n) * (x * error).sum()
    db = (-2/n) * error.sum()

    return dm, db
```

```python
#Step 4: Update Parameters

def update_parameters(m, b, dm, db, alpha):
    m = m - alpha * dm
    b = b - alpha * db
    return m, b
```

```python
#Step 5: Model Training

def initialize_parameters():
    m = 0.0
    b = 0.0
    return m, b

def train(x, y, alpha=0.01, iterations=10):

    m, b = initialize_parameters()
    loss_history = []   # store losses

    for i in range(iterations):

        # Prediction
        y_hat = predict(x, m, b)

        # Cost
        loss = compute_cost(y, y_hat)
        loss_history.append(loss)

        # Gradients
        dm, db = compute_gradients(x, y, y_hat)

        # Update
        m, b = update_parameters(m, b, dm, db, alpha)

        if i % 200 == 0:
            print(f"Iteration {i}, Loss = {loss:.4f}")

    return m, b, loss_history
```

```python
# Dataset Loading

import numpy as np
import matplotlib.pyplot as plt

# Dataset [satisfying y = 4x + 3]
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([7, 11, 15, 19, 23], dtype=float)

plt.scatter(x, y)

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Dataset")

plt.show()
```

```python
# Train Model (your already implemented train function)
m, b, loss_history = train(x, y, alpha=0.01, iterations=10)

print("\nFinal Parameters:")
print("m =", m)
print("b =", b)

# ---------------------------------
# First Plot: Loss Minimization
# ---------------------------------
plt.figure()

plt.plot(loss_history)
plt.xlabel("Iterations")
plt.ylabel("Loss (MSE)")
plt.title("Loss Minimization Over Iterations")

plt.show()
```

```python
# ---------------------------------
# Second Plot: Regression Line
# ---------------------------------
plt.figure()

y_pred = m * x + b

plt.scatter(x, y)
plt.plot(x, y_pred)

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression - Best Fit Line")

plt.show()
```

```python
#Step 8: Predict New Value

x_new = int(input())
y_new_pred = predict(x_new, m, b)
print("Prediction for",x_new,":", y_new_pred)
```
