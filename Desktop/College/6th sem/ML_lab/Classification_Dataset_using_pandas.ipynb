Creating a Classification Dataset using Dictionary and Convert it into Pandas DataFrame

Objective: To understand how raw data can be created manually and converted into a structured tabular format using Pandas.

Step-by-Step Explanation

Import the Pandas library.
Create a Python dictionary with 5 rows and 4 columns.
Convert the dictionary into a Pandas DataFrame.
Display the dataset.

```python
import pandas as pd

# Step 1: Create dictionary data
data = {
    'Age': [22, 25, 47, 52, 46],
    'Salary': [25000, 32000, 60000, 80000, 65000],
    'Experience': [1, 2, 10, 15, 9],
    'Class': ['No', 'No', 'Yes', 'Yes', 'Yes']
}
```

```python
# Step 2: Convert dictionary to DataFrame
df = pd.DataFrame(data)
```

```python
# Step 3: Display dataset
print(df)
```
