Load Toy and Real-World Datasets from Scikit-Learn and Perform Statistical Analysis

Objective: To load built-in datasets from Scikit-Learn and analyze them statistically.

Step-by-Step Explanation

Import the required library.
Load a toy dataset (Iris).
Load a real-world dataset (Breast Cancer).
Convert datasets into Pandas DataFrames.
Perform statistical analysis using describe().

```python
import pandas as pd
from sklearn.datasets import load_iris, load_breast_cancer

# Step 1: Load toy dataset (Iris)
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
iris_df.describe()
```

```python
# Step 2: Load real-world dataset (Breast Cancer)
cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
cancer_df.head()
```

```python
cancer_df.describe()
```
