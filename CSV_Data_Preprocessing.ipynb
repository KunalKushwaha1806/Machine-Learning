Load a CSV Dataset and Perform Data Preprocessing

Objective: To clean and preprocess data before applying Machine Learning algorithms.

Step-by-Step Explanation

Load CSV dataset.
Check missing values.
Handle missing values.
Encode categorical variables..
Normalize numerical data.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_csv("Heart.csv")
df
```

```python
# Drop index column
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
# Check missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
print(df.isnull().sum())
```

```python
# Encode categorical columns
encoder = LabelEncoder()
categorical_cols = ['Sex', 'ChestPain', 'RestECG', 'Slope', 'Thal', 'AHD']
```

```python
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
df
```

```python
# Normalize numerical columns
scaler = StandardScaler()
numerical_cols = ['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak', 'Ca']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```
